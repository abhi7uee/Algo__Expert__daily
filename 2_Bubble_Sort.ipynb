# Bubble sort

```python
#ascending order

def bubbleSort(array):
    length_of_array = len(array)
    for counter in range(1, length_of_array, 1):
        for index in range(0, length_of_array-1,1):
            if array[index] > array[index+1]:
                temp = array[index]
                array[index] = array[index+1] 
                array[index+1] = temp
        length_of_array -= 1
    return array
    
array = [6,4,8,2,9]
sorted_array = bubbleSort(array)
print(sorted_array)
        
```

```python
#Worst case scenario: TC: O(n^2) | SC: O(1)
#Average case scenario: TC: O(n^2) | SC: O(1)

#Best case scenario (when the array is already sorted): TC: O(n) | SC: O(1)
#To check whether the array is already sorted or not, add a check.

def bubbleSort(array):
    isSorted = True
    length_of_array = len(array)
    for counter in range(1, length_of_array, 1):
        for index in range(0, length_of_array-1,1):
            if array[index] > array[index+1]:
                temp = array[index]
                array[index] = array[index+1] 
                array[index+1] = temp
                isSorted = False
        if isSorted:
            break
        else:
            length_of_array -= 1
    return array
```

```python
array = [6,4,8,2,9]
#array = [2, 4, 6, 8, 9]
sorted_array = bubbleSort(array)
print(sorted_array)
```
